```python
import numpy as py
import csv as csv
import pandas as pd
```

```python
studentsFile = "../AcademyOfPy/students_complete.csv"
schoolsFile = "../AcademyOfPy/schools_complete.csv"
studentsData = pd.read_csv(studentsFile)
schoolsData = pd.read_csv(schoolsFile)
```

```python
percentPassingMath = sum(i > 70 for i in studentsData['math_score'])/len(studentsData)*100
percentPassingReading = sum(i > 70 for i in studentsData['reading_score'])/len(studentsData)*100
overallPassing = (percentPassingMath + percentPassingReading)/2 

districtSummary = pd.DataFrame({'Total Schools':[len(schoolsData)], 'Total Students':[len(studentsData)], 
                                'Total Budget':[schoolsData['budget'].sum()],
                                'Average Math Score':[studentsData['math_score'].mean()],
                               'Average Reading Score':[studentsData['reading_score'].mean()], 
                                '% Passing Math':[percentPassingMath], 
                                '% Passing Reading':[percentPassingReading],
                                'Overall Passing Rate':[overallPassing]})
districtSummary = districtSummary[['Total Schools','Total Students','Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']]
districtSummary
```

```python
schoolSummary = schoolsData
schoolSummary.drop('School ID', axis = 1, inplace = True)
schoolSummary['Per Student Budget'] = 0
schoolSummary['Average Math Score'] = 0
schoolSummary['Average Reading Score'] = 0
schoolSummary['% Passing Math'] = 0
schoolSummary['% Passing Reading'] = 0
schoolSummary['% Overall Passing Rate'] = 0
schoolSummary.set_index("name", inplace = True)

for i in schoolSummary.index:
    numPassingMath = 0
    numPassingReading = 0
    sumMath = 0
    sumReading = 0
    schoolSummary.loc[i, 'Per Student Budget'] = schoolSummary.loc[i,'budget']/schoolSummary.loc[i,'size']
    
    for j in studentsData.index:
        if studentsData.loc[j, 'school'] == i:
            sumMath += studentsData.loc[j, 'math_score']
            sumReading += studentsData.loc[j, 'reading_score']
            if studentsData.loc[j, 'math_score'] > 70:
                numPassingMath += 1
            if studentsData.loc[j, 'reading_score'] > 70:
                numPassingReading += 1
                
    schoolSummary.loc[i, 'Average Math Score'] = sumMath/schoolSummary.loc[i,'size']
    schoolSummary.loc[i, 'Average Reading Score'] = sumReading/schoolSummary.loc[i,'size']
    schoolSummary.loc[i, '% Passing Math'] = numPassingMath/schoolSummary.loc[i,'size'] * 100
    schoolSummary.loc[i, '% Passing Reading'] = numPassingReading/schoolSummary.loc[i,'size'] * 100
    schoolSummary.loc[i, '% Overall Passing Rate'] = (schoolSummary.loc[i, '% Passing Math'] + schoolSummary.loc[i, '% Passing Reading'])/2
    
schoolSummary
```

```python
schoolSummary.sort_values(by=['% Overall Passing Rate'], ascending = False).head(5)
```

```python
mathByGrade = schoolSummary.drop(schoolSummary.columns[0:9], axis = 1)
mathByGrade['9th'] = mathByGrade['10th'] = mathByGrade['11th'] = mathByGrade['12th'] = 0
mathByGrade.loc['Current Count'] = [0,0,0,0]
for i in mathByGrade.index:
    mathByGrade.loc['Current Count'] = [0,0,0,0]
    currentGrade = 0
    for j in studentsData.index:
        if studentsData.loc[j, 'school'] == i:
            currentGrade = studentsData.loc[j, 'grade']
            mathByGrade.loc[i, currentGrade] += studentsData.loc[j, 'math_score']
            mathByGrade.loc['Current Count', currentGrade] += 1
            
    for col in mathByGrade:
        mathByGrade.loc[i,col] /= mathByGrade.loc['Current Count', col]

mathByGrade.head(len(mathByGrade)-1)
    
```
